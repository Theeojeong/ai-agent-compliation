```python
import requests

BASE_URL = "Http://127.0.0.1:8000"
APP_NAME="travel_advisor_agent"
USER_ID="jaehyeon"
SESSION_ID = "0350d5ee-4082-4c0b-b857-15a0f200a015"
```

```python
response = requests.post(f"{BASE_URL}/apps/{APP_NAME}/users/{USER_ID}/sessions")

print(response.json())
```

```python
message = {
    "appName": APP_NAME,
    "userId": USER_ID,
    "sessionId": SESSION_ID,
    "newMessage": {
        "parts": [
            {
            "text": "북경 여행을 계획 중인데, 그곳의 날씨와 현재 환율을 알려줘.",
            }
        ],
        "role": "user"
    },
    "streaming": True,
}
```

```python
response = requests.post(url=f"{BASE_URL}/run_sse", json=message)

# (response.json())
```

```python
import sseclient
import json

client = sseclient.SSEClient(response)

for event in client.events():
    data = json.loads(event.data)
    content = data.get("content")
    parts = content.get("parts")
    #text = parts[0]['text']
    print(parts)
    print("="*80)
```
